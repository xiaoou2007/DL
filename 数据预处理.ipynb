```python
import pandas as pd
data = pd.read_csv(data_file)
print(data)
```

```python
inputs,outputs = data.iloc[:,0:2],data.iloc[:,2]
inputs = inputs.fillna(inputs.mean(numeric_only=True))
print(inputs)
```

```python
inputs = pd.get_dummies(inputs,dummy_na=True)
#这是一个 Pandas 提供的强大函数，用于将分类变量（Categorical Variables）转换为独热编码。
#原理：它会扫描数据中的每一列。如果发现某列是文本（类别），它就会根据这一列里有多少种不同的取值，生成多少个新的“0/1”列。
#举例：如果 Alley 列有 "Pave" 和 "Grvl" 两种值，它就会生成两列：Alley_Pave 和 Alley_Grvl。如果是 "Pave"，则前者为 1，后者为 0。
print(inputs)
```
