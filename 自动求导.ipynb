```python
import torch
x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True)#等价于 x=torch.arange(4.0,requires_grad=True)
x.grad
#告诉 PyTorch：“这个张量很重要，我要对它求导。”
#准备好接收“导数结果”的容器
```

```python
y = 2 * torch.dot(x,x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

```python
x.grad.zero_()#梯度清零，否则梯度会累加
y = x.sum()
y.backward()
x.grad
```

```python
x.grad.zero_()
y = x * x
y.sum().backward()#加sum()是为了把向量转为标量
x.grad
```

```python
x.grad.zero_()
y = x * x
u = y.detach()#把u当作了常数系数
z = u * x
z.sum().backward()
x.grad == u
```
