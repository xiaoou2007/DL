```python
A = torch.arange(20,dtype=torch.float32).reshape(5, 4)
print(A)
sum_A = A.sum(axis=1,keepdim=True)
print(A/sum_A)
#这个操作实际上是在做行归一化（Row Normalization）。它计算了每一行中，每个元素占该行总和的比例。这在机器学习中常用于将数据转换为概率分布（例如 Softmax 的前置步骤）
```
